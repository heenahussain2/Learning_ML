### Random Smapling and Sampling Bias
    -> A sample is a subset of data from a larger dataset. This larger dataset is called as population.
    -> A population in statistics is not the same as biology, it is a large, defined or sometimes theoritical set of data.
    -> Random Sampling is a process in which each available member of the population being sampled has an equal chance of being chosen
       for the sample at each draw.
    -> There are two types of sampling:
       1. Replacement - In this the observations are put back in the population after every draw for possible fututre selection.
       2. w/o replacement - In this the observations once selected, are unavailable for future draws.

#### NOTE
     -> Data quality often matters more than data quantity, when making an estimate or a model based on the sample
     -> In data science it involves - completeness, consistency of format, cleanliness and accuracy of individual data points.

### Bias
    -> Statistical Bias refers to measurement or sampling errors, that are systematic and produced by measurement or sampling process.
    -> Its basically if you choose a sample which doesnot correctly represent the larger population, is called bias.

### Random Selection (Sampling)
    -> Drawing elements for a sample at random
    -> There are different types of random sampling. The most common one is stratified sampling.
    -> In stratified sampling, the population is divided up into strata( categories) and random samples are taken from each stratum.
    -> The strata with less size is given more weight, to yield equivalent sample sizes.



### Selection Bias
    -> Bias resulting from the way in which observations are selected.
    -> This can occur in two ways:
       1. Data Snooping: Extensive hunting through data in search of something interesting. There is a saying among statisticians - "If you torture the data long enough, sooner or later it will confess."
       2. Vast Search Effect: Bias or non reproducability resulting from repeated data modeling, or modeling data with large number of predictor variables.
       -> If you repeatedly run different models and ask different questions with a large dataset you are bound to find something interesting, or is it the chance outlier?
       -> We can guard against this by using a holdout set, and sometimes more than one holdout set, against which to validate performance.

    -> Other forms of selection bias include non random sampling, cherry-picking data, selection of time intervals that accentuate a particular statistical effect, and stopping experiments when the results look "interesting".

### Regression to the Mean (Reversion to the mean/ mediocrity)
    -> Regression (go back) to the mean refers to a phenomenon involving successive measurements on a given variable: extreme observations tend to be followed by more central ones.
    -> Attaching extreme focus and meaning to an extreme value can lead to selection bias.
